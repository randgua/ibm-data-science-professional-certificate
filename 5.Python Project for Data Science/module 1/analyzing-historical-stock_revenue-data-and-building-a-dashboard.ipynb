```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

def get_tesla_revenue():
    # URL containing Tesla's revenue data
    url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
    
    # Add headers to prevent blocking
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"}
    
    # Get the webpage content
    html_data = requests.get(url, headers=headers).text
    
    # Create BeautifulSoup object
    soup = BeautifulSoup(html_data, 'html.parser')
    
    # Find the revenue table
    tables = soup.find_all('table')
    
    for table in tables:
        if 'Tesla Quarterly Revenue' in str(table):
            tesla_revenue = pd.read_html(str(table))[0]
            break
    
    # Clean the data
    tesla_revenue.columns = ['Date', 'Revenue']
    
    # Remove dollar signs and convert to numeric, handling 'B' for billions
    tesla_revenue['Revenue'] = tesla_revenue['Revenue'].str.replace('$', '').str.replace(',', '')
    tesla_revenue['Revenue'] = pd.to_numeric(tesla_revenue['Revenue'], errors='coerce')
    
    # Remove rows with NaN values
    tesla_revenue.dropna(inplace=True)
    
    return tesla_revenue

# Get the data
tesla_revenue = get_tesla_revenue()

# Display the last 5 rows
print("\nLast 5 rows of Tesla Revenue Data:")
print(tesla_revenue.tail())
```

```python
import pandas as pd

def get_tesla_revenue_yahoo():
    # URL for Tesla's financials on Yahoo Finance
    url = "https://finance.yahoo.com/quote/TSLA/financials?p=TSLA"
    
    # Read all tables from the URL
    # Yahoo Finance tables are well-structured HTML
    tables = pd.read_html(url)
    
    # The income statement is typically the first table
    income_statement = tables[0]
    
    # Set the first column as index
    income_statement.set_index(income_statement.columns[0], inplace=True)
    
    # Get the Total Revenue row
    revenue_data = income_statement.loc['Total Revenue']
    
    # Convert to DataFrame and transpose
    revenue_df = pd.DataFrame(revenue_data).T
    
    # Reset index to make dates a column
    revenue_df = revenue_df.reset_index()
    revenue_df.columns = ['Date', 'Revenue']
    
    # Clean the Revenue column (remove thousands separator and convert to numeric)
    revenue_df['Revenue'] = revenue_df['Revenue'].replace(',', '', regex=True)
    revenue_df['Revenue'] = pd.to_numeric(revenue_df['Revenue'], errors='coerce')
    
    # Sort by date
    revenue_df = revenue_df.sort_values('Date', ascending=True)
    
    return revenue_df

try:
    # Get the data
    tesla_revenue = get_tesla_revenue_yahoo()
    
    # Display the last 5 rows
    print("\nLast 5 rows of Tesla Revenue Data:")
    print(tesla_revenue.tail())
    
except Exception as e:
    print(f"An error occurred: {str(e)}")
    print("Note: Yahoo Finance might block automated access. Consider using an API or adding headers.")
```
